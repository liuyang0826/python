```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['figure.dpi'] = 200
```

```python
img = cv2.imread("u103.png")
plt.imshow(img, "gray")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

res = cv2.morphologyEx(gray, cv2.MORPH_OPEN, np.ones((9,9),np.uint8), 30)
gray = cv2.medianBlur(gray,3)
res = cv2.Canny(res,40,220)
contours, h = cv2.findContours(res, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
# plt.imshow(contours, "gray")

# draw_img = img.copy()
# ret = cv2.drawContours(draw_img, contours, -1, (0, 0, 255), 1)
plt.imshow(res, "gray")
```
